Graph isomorphism is a concept in graph theory used to determine whether two graphs are structurally identical. Two graphs are said to be isomorphic if there is a one-to-one correspondence between their vertices and edges such that the adjacency relationships are preserved.

```python
import networkx as nx
```

```python
# Create the first graph
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4)])

# Create the second graph
G2 = nx.Graph()
G2.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D')])
```

```python
# Check if the graphs are isomorphic
GM = nx.algorithms.isomorphism.GraphMatcher(G1, G2)
print("Graphs are isomorphic:", GM.is_isomorphic())
```

# Summary
Graph isomorphism helps in identifying structural similarities between graphs, which is useful in various applications, from pattern recognition to network analysis.

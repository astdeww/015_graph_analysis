# Community Detection 
The key concept in graph analysis is to be used to identify clusters or groups of nodes within a network that are more densely connected to each other than to the rest of the network. This has real-life applications in social networks, biology, marketing, and more.

## Real-Life Example: Social Network Analysis
Imagine a social network where individuals (nodes) are connected if they interact frequently. Community detection can help identify groups of people who interact more within their group than with others. This can be useful for understanding social dynamics, targeted marketing, or identifying influential groups.

```python
import networkx as nx
import matplotlib.pyplot as plt
import community as community_louvain
```

```python
# Create a sample graph
G = nx.Graph()

# Add nodes and edges
nodes = ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Judy']
edges = [
    ('Alice', 'Bob'), ('Alice', 'Charlie'), ('Bob', 'Charlie'), 
    ('David', 'Eve'), ('Eve', 'Frank'), ('David', 'Frank'),
    ('Grace', 'Heidi'), ('Heidi', 'Ivan'), ('Ivan', 'Judy'), ('Grace', 'Judy')
]

G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
# Apply Louvain method for community detection
partition = community_louvain.best_partition(G)
```

```python
# Print community assignment
print("Community assignment for each node:")
for node, community in partition.items():
    print(f"Node {node} is in community {community}")
```

```python
# Draw the graph with community detection
pos = nx.spring_layout(G, seed=7)
plt.figure(figsize=(10, 8))
```

```python
# Draw nodes with colors according to community
cmap = plt.get_cmap('viridis', len(set(partition.values())))
nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=500, cmap=cmap, node_color=list(partition.values()))
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.title('Community Detection in Social Network')
plt.colorbar(plt.cm.ScalarMappable(cmap=cmap), label='Community')
plt.show()
```
